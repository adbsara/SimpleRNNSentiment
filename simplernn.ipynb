```python
max_features=10000 ##vocabulary size
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)

# Print the shape of the data
print(f'Training data shape: {X_train.shape}, Training labels shape: {y_train.shape}')
print(f'Testing data shape: {X_train.shape}, Testing labels shape: {y_test.shape}')
```

```python
from tensorflow.keras.preprocessing import sequence

max_len=500

X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test = sequence.pad_sequences(X_test, maxlen=max_len)
X_train
```

```python
X_train[0]
```

```python
## Train Simple RNN
model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len)) ## Embedding Layers
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## Create an instance of EarlyStoppping Callback
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
## Train the model with early sstopping
history=model.fit(
    X_train,y_train,epochs=10,batch_size=32,
    validation_split=0.2,
    callbacks=[earlystopping]
)
```

```python
## Save model file
model.save('simple_rnn_imdb.h5')
```

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
from sklearn.model_selection import GridSearchCV
import numpy as np

# تعریف کلاس سفارشی KerasClassifier
class CustomKerasClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self, build_fn=None, neurons=32, layers=1, epochs=50, batch_size=32, **kwargs):
        self.build_fn = build_fn
        self.neurons = neurons
        self.layers = layers
        self.epochs = epochs
        self.batch_size = batch_size
        self.kwargs = kwargs
        self.model = None

    def build_model(self):
        if self.build_fn is None:
            raise ValueError("No build function provided for the model.")
        return self.build_fn(neurons=self.neurons, layers=self.layers)

    def fit(self, X, y, **fit_params):
        self.model = self.build_model()
        y = np.array(y)  # اطمینان از اینکه y به شکل آرایه numpy است
        self.model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, **fit_params)
        return self

    def predict(self, X):
        return (self.model.predict(X) > 0.5).astype("int32").flatten()

    def score(self, X, y):
        predictions = self.predict(X)
        return np.mean(predictions == y)

    def get_params(self, deep=True):
        return {
            "build_fn": self.build_fn,
            "neurons": self.neurons,
            "layers": self.layers,
            "epochs": self.epochs,
            "batch_size": self.batch_size,
        }

    def set_params(self, **params):
        for key, value in params.items():
            setattr(self, key, value)
        return self

    def get_tags(self):
        return {"estimator_type": "classifier"}  # مشخص کردن نوع تخمین

```

```python
def create_model(neurons=32, layers= 1):
    model = Sequential()
    model.add(Dense(neurons, activation= 'relu', input_shape=(X_train.shape[1],)))

    for _ in range(layers-1):
        model.add(Dense(neurons, activation='relu'))

    model.add(Dense(1, activation="sigmoid"))


    model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])
    
    return model
```

```python

model = CustomKerasClassifier(build_fn=create_model, epochs=50, batch_size=10, verbose=0)

# تعریف پارامترها برای GridSearchCV
param_grid = {
    'neurons': [16, 32, 64, 128],
    'layers': [1, 2, 3],
    'epochs': [50, 100],
    'batch_size': [10, 20]
}

# تعریف GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)

# اجرای GridSearchCV
grid_result = grid.fit(X_train, y_train)

# نمایش بهترین پارامترها
print("Best parameters found: ", grid_result.best_params_)

# امتیاز بهترین مدل
print("Best score: ", grid_result.best_score_)
```
